```python
#scatter plot
plot_df.plot(kind="scatter", x="Pifzer/Moderna Total Doses", y="Confirmed", grid=True, figsize=(8,8),
              title="Total Doses Allocated vs Total Confirmed Cases")
#correlation
correlation = st.pearsonr(plot_df['Pifzer/Moderna Total Doses'], plot_df['Confirmed'])
print(f"The correlation between both factors is {round(correlation[0],2)}")
plt.show()
```

```python
plot_df.plot(kind="scatter", x="Pifzer/Moderna Total Doses", y="Density", grid=True, figsize=(8,8),
              title="Total Doses Allocated vs Population Density")
#correlation
correlation = st.pearsonr(plot_df['Pifzer/Moderna Total Doses'], plot_df['Density'])
print(f"The correlation between both factors is {round(correlation[0],2)}")

plt.show()
```

```python
plot_df.plot(kind="scatter", x="Pifzer/Moderna Total Doses", y="Pop", grid=True, figsize=(8,8),
              title="Total Doses Allocated vs Population")

#correlation
correlation = st.pearsonr(plot_df['Pifzer/Moderna Total Doses'], plot_df['Pop'])
print(f"The correlation between both factors is {round(correlation[0],2)}")
plt.show()
```

```python
plot_df.plot(kind="scatter", x="Confirmed", y="Pop", grid=True, figsize=(8,8),
              title="Confirmed Cases vs Population")

#correlation
correlation = st.pearsonr(plot_df['Confirmed'], plot_df['Pop'])
print(f"The correlation between both factors is {round(correlation[0],2)}")
plt.show()
```
