```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Display the data table for preview
unemployment_data_us_state_path = "unemployment_data_us_state.csv"
Population_Density_path = "Population_Density.csv"
Confirmed_Cases_By_State_path = "Confirmed_Cases_By_State.csv"
unemployment_total_data_us_state_path = "unemployment_total_data_us_state.csv"

# Read the mouse data and the study results
unemployment_data_us_state = pd.read_csv(unemployment_data_us_state_path)
Population_Density = pd.read_csv(Population_Density_path)
Confirmed_Cases_By_State = pd.read_csv(Confirmed_Cases_By_State_path)
unemployment_total_data_us_state = pd.read_csv(unemployment_total_data_us_state_path)

# Combine the data into a single dataset
merge_data_df = pd.merge(unemployment_data_us_state, Confirmed_Cases_By_State, on="State")

# Display the data table for preview
merge_data_df.head()
```

### Merge files

```python
#read the data file
merge_df2 = pd.merge(unemployment_total_data_us_state, Population_Density, on="State")

# print clean data frame
merge_df2.head()
```

```python
# Combine the data into a single dataset
merge_data_df3 = pd.merge(merge_data_df, merge_df2, on="State")

# Display the data table for preview
merge_data_df3.head()
```

### Plot the Amount of Covid-19 cases per State

```python
merge_data_df['Confirmed'] = merge_data_df['Confirmed'].str.replace(",", "")
```

```python
merge_data_df['Confirmed']= merge_data_df['Confirmed'].astype(int)
x_axis = np.arange(len(merge_data_df))

#set size
plt.figure(figsize=(40,10))

#set color, rotations of label and x axis
plt.bar(x_axis, merge_data_df["Confirmed"], color='g')
plt.xticks(x_axis, merge_data_df["State"], rotation="vertical")

#Set Title and label
plt.title('Total Covid_19 cases recored in Each State')
plt.ylabel('Cases (Millions)')
plt.xlabel('State')

#print
plt.show()
```

### Plot the Change in Rank of Unemployment by State from 2019 to 2020

```python
# Generate a bar plot showing the Rank of each state.
merge_data_df.plot(kind="scatter", y="Change", x="State", facecolor="green", 
                    edgecolors="black", grid=True, title="Unemployment Rank for Each State")

#rotate labels 
plt.xticks(rotation=90)

#set hline at zero
plt.axhline(y=0, color='r', linestyle='-')

#print
plt.show()
```

```python
x_axis = np.arange(len(merge_data_df))

#set size
plt.figure(figsize=(25,5))

#set color, rotations of label and x axis
x_axis = np.arange(len(merge_data_df["Change"]))
plt.bar(x_axis, merge_data_df["Change"], color='g')  
plt.xticks(x_axis, merge_data_df["State"], rotation="vertical")


#set hline at zero
plt.axhline(y=0, color='r', linestyle='-')

#Set Title and label
plt.title('Change in Rank of Unemployment in Each State from 2019 - 2020')
plt.ylabel('Change in Rank')
plt.xlabel('State')

#print
plt.show()
```

### Plot the average amount of Unemployed in 2019 and 2020. Then show the Average of the two years

```python
# remove commas
merge_data_df3['Confirmed'] = merge_data_df3['Confirmed'].str.replace(",", "")
merge_data_df3['Average Unemployed 2019'] = merge_data_df3['Average Unemployed 2019'].str.replace(",", "")
merge_data_df3['Average Unemployed 2020'] = merge_data_df3['Average Unemployed 2020'].str.replace(",", "")
merge_data_df3['Total Average'] = merge_data_df3['Total Average'].str.replace(",", "")

#print
merge_data_df3.head()
```

```python
# Change dtypes to int
merge_data_df3['Average Unemployed 2019']= merge_data_df3['Average Unemployed 2019'].astype(int)
merge_data_df3['Average Unemployed 2020'] = merge_data_df3['Average Unemployed 2020'].astype(int)
merge_data_df3['Total Average'] = merge_data_df3['Total Average'].astype(int)
# merge_data_df3.dtypes
```

```python
#setting up bar chart
ax=merge_data_df3['Total Average'].plot(color = 'red',linewidth=2.0, use_index = True)
leg=plt.legend()
merge_data_df3[['Average Unemployed 2019','Average Unemployed 2020']].plot(kind = 'bar',stacked = True, 
color = ['grey','navy'], figsize=(20,10), ax=ax)
leg2=plt.legend()


ax.set_xticks(merge_data_df3.index)
ax.set_xticklabels(merge_data_df3['State'], rotation="vertical")

plt.legend(leg.get_lines()+leg2.get_patches(), [text.get_text() for text in leg.get_texts()+leg2.get_texts()])

#print
plt.show()
```

```python
#add the total unemployed
sum_column = merge_data_df3['Average Unemployed 2019'] + merge_data_df3['Average Unemployed 2020']
merge_data_df3["Total Unemployed"] = sum_column

#print
merge_data_df3.head()
```

### Make New Data Frame and Chart the Top 5 States and the lowest 5 States for Confirmed Cases and Unemployed 

```python
#Plot and set up new data frame
data =merge_data_df3.set_index('State')
state = merge_data_df3['State']
cases = merge_data_df3['Confirmed']
unemployed = merge_data_df3['Total Unemployed']
summary = pd.DataFrame({'State':state,'Confirmed':cases, 'Total Unemployed':unemployed})

#print results
summary.head()
```

```python
merge_data_df3['Confirmed']= merge_data_df3['Confirmed'].astype(int)
x_axis = np.arange(len(merge_data_df3))

#set size
plt.figure(figsize=(25,8))

#set color, rotations of label and x axis
plt.bar(x_axis, merge_data_df3["Total Unemployed"], color='g')
plt.xticks(x_axis, merge_data_df3["State"], rotation="vertical")

#Set Title and label
plt.title('Total Unemployed in 2019 and 2020 in Each State')
plt.ylabel('Cases (Millions)')
plt.xlabel('State')

#print
plt.show()
```

```python
#putting in order by the highest
summary = summary.sort_values(by="Confirmed", ascending=False)

#dropping everything but the top 10 states
summary = summary.drop([30,14,22,21,0,23,25,18,5,36,17,20,15,47,46,3,16,28,24,6,27,32,12,48,2,34,26,7,29,13,19,11,35,43,41,10,38,39,9], axis=0)

#Show output
summary
```

```python
#Plot Bar Chart
ax = summary.plot.bar(x='State', figsize=(15,8), rot=0)

#Set Title and label
plt.title('Top 5 States and the lowest 5 States of Unemployment VS Confirmed Cases')
plt.ylabel('(Millions)')
plt.xlabel('State')
```
