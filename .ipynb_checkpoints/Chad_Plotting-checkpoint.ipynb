```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
```

```python
# open dataframe to work with
# file path
csv = 'Cleaned_Data/Final.csv'
#read csv
plot_df = pd.read_csv(csv)
plot_df.head()
```

```python
#scatter allocated dosage vs confirmed cases
#axis variable
x_values = plot_df['Pifzer/Moderna Total Doses']
y_values = plot_df['Confirmed']
#liner regression equation
slope, intercept, rvalue, pvalue, stderr = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")


#plot
plt.scatter(x_, y_axis, marker="o", facecolors="red", edgecolors="black",
            #alpha=0.75)
# plot info
# plt.title("Total Doses Allocated vs Total Confirmed Cases")
# plt.xlabel("Doses Alloacted ")
# plt.ylabel("Confimred Cased")

#plt.show()
```
